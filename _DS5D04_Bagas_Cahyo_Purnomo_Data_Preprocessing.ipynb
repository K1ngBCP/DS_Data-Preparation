karena terlihat dengan jelas adanya titikdiatas atau dibawah kotak maka data ini terindikasi memiliki outlier  dan harus dihandling. disini saya menggunakn Z-score sebagai metode handling outliersnya dimana yang sya handling adalah kolom 'trestbps', 'chol', 'fbs', 'thalach', 'oldpeak', 'ca', 'thal'

```python
# Z-Score
columns = ['trestbps', 'chol', 'fbs', 'thalach', 'oldpeak', 'ca', 'thal']

for i in columns:
    upper_limit = Dataku[i].mean() + 3*Dataku[i].std()
    lower_limit = Dataku[i].mean() - 3*Dataku[i].std()

    Dataku[i] = np.where(
       Dataku[i]>upper_limit,
        upper_limit,
        np.where(
            Dataku[i]<lower_limit,
            lower_limit,
            Dataku[i]
        )
   )
```

```python
# Fungsi untuk mengkategorikan umur
def categorize_age(age):
    if age <= 5:
        return 0 # Balita
    elif 5 < age <= 11:
        return 1 # anak anak
    elif 12 <= age < 25:
        return 2 #Remaja
    elif 26 <= age < 60:
        return 3 #Dewasa
    else:
        return 4 # Lansia

Dataku['age_category'] = Dataku['age'].apply(categorize_age)
```

Disini saya membuat kategorisasi umur jadi saya menggolongkannya menjadi 5 kategori dan menampilkannya menjadi 0,1,2,3 atau 4 dimana pengkategoriannya saya menggunakan acuan data di bawah ini. Saya menyimpannya di kolom age_category

Acuan : https://gaya.tempo.co/read/1724197/kategori-umur-balita-remaja-dan-dewasa-menurut-kemenkes-jangan-salah

```python
def tekanan_darah(trestbps):
    if trestbps <= 90:
        return 0  # Hipotensi
    elif 91 <= trestbps <= 139:
        return 1  # Normal
    else:
        return 2  # Hypertension

Dataku["trestbps_category"] = Dataku["trestbps"].apply(tekanan_darah)
```

lalu disini saya membuat fitur untuk menampilkan normaltidaknya tekanan darah secara umum dimana sayamengkategorikannya menjadi 3 kelompok dan saya meletakkannya di kolom trestbps_category dengan mengambil data dibawah ini

Acuan : https://www.siloamhospitals.com/en/informasi-siloam/artikel/apa-itu-hipotensi

```python
def categorize_chol(chol):
    if chol < 200:
        return 0  # Normal
    elif 200 <= chol <= 239:
        return 1  # Kurang Normal/batas normal
    else:
        return 2  # Tidak normal


Dataku["chol_category"] = Dataku["chol"].apply(categorize_chol)
```

Disini saya mengkategorisasikan kalori kedalam 3 kelompok danmeletakkannya di chol_category dengan acuan data pada link dibawah

Acuan : https://www.siloamhospitals.com/informasi-siloam/artikel/kadar-kolesterol-normal

```python
def cholxage(age,chol):
  if age <= 19 :
    if chol < 170 :
      return 0 #Normal
    else :
      return 1 #tidak normal
  elif age > 19 :
    if 125 <= chol <= 200 :
      return 0 #Normal
    else :
      return 1 #tidak normal

Dataku["chol_catxage"] = Dataku.apply(lambda row: cholxage(row['age'], row['chol']), axis=1)
```

Disini saya mengkategorikan kalori yang normal berdasarkan umur dimana ada 2 kategori umur yakni 19 kebawah dan 20 keatas dimana acuan datanya ada di link dibawah

Acuan : https://health.kompas.com/read/23B22160000968/berapa-kadar-kolesterol-normal-berdasarkan-usia?page=all

```python
def trestbpsxage(age,trestbps):
  if age <= 2 :
    if 85 <= trestbps <= 113 :
      return 0 #Normal
    else :
      return 1 #tidak normal
  elif 3<= age <= 5 :
    if 91 <= trestbps <= 120 :
      return 0 #Normal
    else :
      return 1 #tidak normal
  elif 6<= age <= 12 :
    if 96 <= trestbps <= 131 :
      return 0 #Normal
    else :
      return 1 #tidak normal
  elif 13<= age <= 18 :
    if 112 <= trestbps <= 128 :
      return 0 #Normal
    else :
      return 1 #tidak normal
  elif 19<= age <= 59 :
    if 90 <= trestbps <= 120 :
      return 0 #Normal
    else :
      return 1 #tidak normal
  elif age >= 60 :
    if 90 <= trestbps <= 120 :
      return 0 #Normal
    else :
      return 1 #tidak normal

Dataku["trestbps_age"] = Dataku.apply(lambda row: cholxage(row['age'], row['trestbps']), axis=1)
```

lalu disini saya menghitung trestbps berdasarkan age atau usia untuk mengetahui dalam usia sekian apakah tekanan darah pada dataset tersebut sudah normal dimana ada 6 kategori umur. hasil nya diletakkan di kolom trestbps_age dengan acuan data dibawah ini

Acuan : https://www.siloamhospitals.com/informasi-siloam/artikel/mengenal-tekanan-darah-normal#mcetoc_1h45ndmcigb4

```python
def calculate_mhr(age):
    # Rumus perkiraan MHR: 220 - Usia
    estimated_mhr = 220 - age
    return estimated_mhr

Dataku["calculate_mhr"] = Dataku["age"].apply(calculate_mhr)
```

Disini saya menggunakan perhitungan mhr sederhana dengan cara 220- umur adalah mhr, lalu saya meletakkannya di kolom calculate_mhr

Acuan :https://www.cdc.gov/physicalactivity/basics/measuring/heartrate.htm#:~:text=You%20can%20estimate%20your%20maximum,beats%20per%20minute%20(bpm).

```python
def mhrxthalac (calculate_mhr,thalach):
  if calculate_mhr == thalach :
    return 0 # Normal
  else :
    return 1 # Tidak Normal

Dataku["mhrxthalac"] = Dataku.apply(lambda row: mhrxthalac(row['calculate_mhr'], row['thalach']), axis=1)
```

lalu setelah mengetahui mhrnya disini saya mengkategorisasikannya dengan mengecek apakah mhr sama dengan thalac jika sama berarti normal dilambankan dengan 0 dan jika tidak dilambangkan 1 yang artinya tidak normal

```python
# Bobot untuk masing-masing faktor risiko dan parameter (bebas untuk disesuaikan)
weight_age = 0.2
weight_gender_male = 0.2  # Bobot lebih tinggi untuk pria
weight_gender_female = 0.1  # Bobot lebih rendah untuk wanita
weight_trestbps = 0.2
weight_chol = 0.4
weight_restecg = 0.1

# Fungsi untuk menghitung skor keseluruhan
def calculate_heart_health_score(row):
    # Menggunakan bobot yang berbeda berdasarkan jenis kelamin
    if row['sex'] == '1':
        gender_weight = weight_gender_male
    else:
        gender_weight = weight_gender_female

    score = (
        weight_age * (1 - row['age'] / 100) +
        gender_weight * (1 - row['sex'] / 2) +
        weight_trestbps * (1 - row['trestbps'] / 200) +
        weight_chol * (1 - row['chol'] / 300) +
        weight_restecg * row['restecg']
    )
    return score

Dataku['heart_health_score'] =Dataku.apply(calculate_heart_health_score, axis=1)
```

lalu disini saya menggunakan logika saya sendiri untuk memberikan bobot beberapa kolomuntuk menjadi pembanding (dibantu dengan sumber internet yang ada dibawah), misalnya saja laki laki lebih mudah terserang penyakit jantung (dari sumber internet) maka bobotnya akan lebih besar dari wanita dimana saya meletakkannya di kolom heart_health_score

Acuan : https://ejurnal.esaunggul.ac.id/index.php/Inohim/article/download/2120/1835  
https://www.halodoc.com/artikel/benarkah-jantung-koroner-lebih-sering-menyerang-pria

```python
# Kategorisasi skor kesehatan jantung
def categorize_health_score(score):
    if score >= 0.7:
        return 0 #Baik
    elif 0.4<= score < 0.7:
        return 1 #Sedang
    else:
        return 2 #Buruk

Dataku['health_category'] = Dataku['heart_health_score'].apply(categorize_health_score)
```

lalu setelah itu saya mengkategorikannya dimana jika dibawah 0,4 dianggap buruk yang disimbolkan 2 antara 0,4 sampai 0,7 dianggap sedang disimbolkan 1 dan diatas 0,7 dianggap baik dan disimbolkan 0

```python
Dataku['ecg_thal_index'] = Dataku['restecg'] * Dataku['thal']
```
